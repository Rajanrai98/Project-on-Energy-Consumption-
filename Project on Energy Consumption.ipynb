```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
data =  data.apply(le.fit_transform)
```

```python
data
```

```python
data.isnull().sum()
```

```python
x = data.drop(["EnergyConsumption"] , axis = 1)
y = data['EnergyConsumption']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
x_train , x_test ,y_train , y_test = train_test_split(x , y , test_size = 0.20 , random_state = 42)
```

```python
x_train.shape , x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
pd.DataFrame({"y_test":y_test , "y_pred": y_pred})
```

```python
mse = np.mean((y_test-y_pred)**2)
print("mean_squared_value:",mse)
```

```python
rmse = np.sqrt(mse)
print("rmse_linear:",rmse)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model_tree = DecisionTreeRegressor(criterion = "friedman_mse" , max_depth = 3)
```

```python
model_tree.fit(x_train , y_train)
```

```python
y_pred_tree = model_tree.predict(x_test)
```

```python
y_pred_tree
```

```python
pd.DataFrame({"y_test":y_test , "y_pred_tree":y_pred_tree})
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize = (24,12))
plot_tree(model_tree , filled = True)
```

```python
mse = np.mean((y_test - y_pred_tree)**2)
print("mse:",mse)
```

```python
rmse = np.sqrt(mse)
print("rmse_tree:",rmse)
```

```python
model_forest = RandomForestRegressor(n_estimators = 100 , max_features = 5)
```

```python
model_forest.fit(x_train , y_train)
```

```python
y_pred_forest = model_forest.predict(x_test)
print(y_pred_forest)
```

```python
pd.DataFrame({"y_test":y_test , "y_pred_forest":y_pred_forest})
```

```python
mse = np.mean((y_test-y_pred_forest)**2)
print("mean_squared_error:",mse)
```

```python
rmse = np.sqrt(mse)
print("rmse_random:",rmse)
```

```python
from keras.layers import Dense
from keras.optimizers import Adam
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Dense(12 ,input_dim = 10 , activation = "relu" ))
model.add(Dense(22 , activation = "relu"))
model.add(Dense(14 , activation = "relu"))
model.add(Dense(6 , activation = "relu"))
model.add(Dense(1 , activation = "linear"))
```

```python
history = model.fit(x_train , y_train , batch_size = 32 , epochs = 200 , validation_split = 0.20)
```
